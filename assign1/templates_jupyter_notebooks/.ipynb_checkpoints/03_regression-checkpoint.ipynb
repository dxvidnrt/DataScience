# Notebook for Part 3 - Regression

```python
# Import Matplotlib and Seaborn for visualization
import matplotlib.pyplot as plt

# Import Pandas and NumPy functionality for editing data
import pandas as pd
import seaborn as sns

# Importing IPython display for visualizing intermediate results in the notebook
from IPython.display import display

# Import Regression
from sklearn.linear_model import LogisticRegression
```

## 3.a) Question: Create a logistic regression model and give weights

```python
# load training data
train_set = pd.read_csv('../heart_disease_data/train_dataset.csv')

# define features
descriptive_features = train_set[['thalch', 'age']]
target_feature = train_set['num']

# create logistic regression model
regressor = LogisticRegression(multi_class='ovr')
regressor.fit(descriptive_features, target_feature)

# extract weights and intercepts
weights = regressor.coef_
intercepts = regressor.intercept_

print(f"""Weights:
{weights}""")
print(f"Intercepts: {intercepts}")
```

## 3.b) Question: Compute error

```python
# load test data 
test_set = pd.read_csv('../heart_disease_data/test-dataset-oh.csv')

# define features
test_desc_features = test_set[['thalch', 'age']]
test_target_features = test_set['num']

# calculate accuracy score of the regression model
score = regressor.score(test_desc_features, test_target_features)
print(f"The regression model has a mean accuracy of {score}.\nThis means, that around {int(score*100)}% of the target features are predicted correctly.")

# describe accuracy score for random prediction
num_classes = len(test_set['num'].unique())
print(f"""There are {num_classes} different classes for the target feature 'num', predicting random would result in an average mean accuracy of {1/num_classes}.
Therefore, the achieved accuracy is better than random guessing.""")
```

## 3.c) Question: Plot thalch over age

```python
fig = plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='thalch', hue='num', data=train_set, palette='coolwarm', s=100)

# adding title and labels
plt.title('Age vs Thalach (Heart Disease Presence/Absence)', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Max Heart Rate (Thalach)', fontsize=12)

# display legend
plt.legend(title='Heart Disease (num)', loc='best')

# show the plot
plt.show()

# print answers
print("The performance of the logistic regression model is not that good, because the data is not linearly separable.")
print("""To improve the performance of the logistic model, one could use polynomial feature extension to increase the dimension of the
data and separate the data non-linearly or normalize both descriptive features for better comparison of different units.""")
```
